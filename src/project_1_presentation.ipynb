***Places***

```python
#Downloading interesting places data
TABLE_ID4 = "tblDuWR3nXGmk9OHl"

headers = {"Authorization" : f"Bearer {API_KEY}",
           "Content-Type"  : "application/json"}

airtable_base_url = "https://api.airtable.com/v0"

params = {}

places_table_endpoint = f"{airtable_base_url}/{BASE_ID}/{TABLE_ID4}"

places_data_list = list()

while params.get("offset") != None or len(places_data_list) == 0:
    
    response = requests.get(url = places_table_endpoint, headers = headers, params = params)
    
    print(f"response: {response.status_code}")
    print(f"endpoint: {response.url}")
    
    data = response.json()
    
    offset = data.get("offset")
    
    params["offset"] = offset
    
    places_data_list.extend(data["records"])
```

```python
#Creating the dataframe of places data
places_list = list()
for i in places_data_list:
    places_list.append(i['fields'])

places_data = pd.DataFrame.from_records(places_list)
```

```python
places_data
```

```python
#Visualization of Bern map with the interesting places
sf_map1 = folium.Map(location = [total_countries['bern']['lat'], total_countries['bern']['lon']], zoom_start = 10, tiles = "OpenStreetMap")

incidents1 = plugins.MarkerCluster()

for lat, lon, name in zip(places_data[places_data['Country'] == 'CH']["Latitude"], places_data[places_data['Country'] == 'CH']["Longitude"], places_data["Name"]):
    
    incidents1.add_child(folium.Marker(location = [lat, lon],
                                      popup    = name))
    
sf_map1.add_child(incidents1)

sf_map1
```
